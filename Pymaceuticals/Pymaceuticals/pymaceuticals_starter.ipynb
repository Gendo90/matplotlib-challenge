## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
print(mouse_metadata)
print(study_results)

combined_data = study_results.merge(mouse_metadata, on="Mouse ID")

# Display the data table for preview
combined_data
```

```python
# Checking the number of mice.

#check number of mice in original data set
print(len(study_results["Mouse ID"].unique()))
print(len(mouse_metadata["Mouse ID"].unique()))

#show that the number of mice in the combined data set is the same
print(len(combined_data["Mouse ID"].unique()))
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

#first show how many values there are for each timepoint - that is, how many mice have a given number of timepoints?
grouped_by_id = study_results.groupby("Mouse ID")
check_duplicates = grouped_by_id.count()

check_duplicates = check_duplicates["Timepoint"]
check_duplicates

per_value = check_duplicates.value_counts()
print(per_value)

#group all the mice by id and then by timepoint, so that we can count how many timepoints for each mouse there are
#to see if there are duplicate timepoints for the same mouse (e.g. the count is greater than one for at least one timepoint
#for a particular mouse)
timepoint_duplicates = study_results.groupby(["Mouse ID", "Timepoint"]).count()
timepoint_duplicates =timepoint_duplicates.drop(columns="Tumor Volume (mm3)")
timepoint_duplicates.columns = ["Timepoint Count"]
print(timepoint_duplicates)
timepoint_duplicates = timepoint_duplicates.loc[timepoint_duplicates["Timepoint Count"]>1]


timepoint_duplicates.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.

#this summary statistics table shows data for the mice at the beginning of the study, at timepoint=0

#filter table so that it only shows the remaining mice and tumor values at timepoint 45
start_mice = cleaned_combined_data.loc[cleaned_combined_data["Timepoint"]==0]
print(f"There are {len(start_mice)} mice at the start of the study.")

index_values = ["Mean", "Median", "Variance", "Standard Deviation", "SEM"]

#write a function that computes the required values for each drug regimen and puts it in a series
def statsForTreatment(treatment, dataset):
    treatment_subset = dataset.loc[dataset["Drug Regimen"]==treatment]
    #get mean tumor volume
    tumor_mean = treatment_subset["Tumor Volume (mm3)"].mean()
    #get median tumor volume
    tumor_median = treatment_subset["Tumor Volume (mm3)"].median()
    #get variance in tumor volume
    tumor_variance = treatment_subset["Tumor Volume (mm3)"].var()
    #get standard deviation in tumor volume
    tumor_st_dev = treatment_subset["Tumor Volume (mm3)"].std()
    #get sem in tumor volume
    tumor_sem = treatment_subset["Tumor Volume (mm3)"].sem()
    
    output_df = pd.DataFrame({treatment:[tumor_mean, tumor_median, tumor_variance, tumor_st_dev, tumor_sem]}, index=index_values)
    return output_df

all_treatments = list(cleaned_combined_data["Drug Regimen"].unique())
print(all_treatments)

summary_stats_iterative = statsForTreatment(all_treatments[0], cleaned_combined_data)
summary_stats_iterative.index.rename = "Stat Variables"

for treatment in all_treatments[1:]:
    new_stats = statsForTreatment(treatment, start_mice)
    new_stats.index.rename = "Stat Variables"
    summary_stats_iterative = summary_stats_iterative.merge(new_stats, left_index = True, right_index = True)
    
summary_stats_iterative
    
    
    
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function

#this summary statistics table shows the tumor volume for the mice by treatment at the end of the study (timepoint = 45)
final_mice = cleaned_combined_data.loc[cleaned_combined_data["Timepoint"]==45]
print(f"There are {len(final_mice)} mice left at the end of the study.")


#group by drug regimen to get summary stats
grouped_treatments = final_mice.groupby("Drug Regimen")

#calculate the mean across all treatments
final_treatment_tumor_means = pd.DataFrame(grouped_treatments.mean()["Tumor Volume (mm3)"])
final_treatment_tumor_means.columns = ["Mean"]

#set the summary stats dataframe to be used for the final mice statistics
treatment_outcome_summary_stats = final_treatment_tumor_means

#calculate the median across all treatments
final_treatment_tumor_medians = pd.DataFrame(grouped_treatments.median()["Tumor Volume (mm3)"])
final_treatment_tumor_medians.columns = ["Median"]

#calculate the variance across all treatments
final_treatment_tumor_variances = pd.DataFrame(grouped_treatments.var()["Tumor Volume (mm3)"])
final_treatment_tumor_variances.columns = ["Variance"]

#calculate the standard deviation across all treatments
final_treatment_tumor_st_dev = pd.DataFrame(grouped_treatments.std()["Tumor Volume (mm3)"])
final_treatment_tumor_st_dev.columns = ["Standard Deviation"]

#calculate the sem across all treatments
final_treatment_tumor_sem = pd.DataFrame(grouped_treatments.sem()["Tumor Volume (mm3)"])
final_treatment_tumor_sem.columns = ["SEM"]

#add all the calculated columns to the final summary stats table
treatment_outcome_summary_stats = treatment_outcome_summary_stats.merge(final_treatment_tumor_medians, on="Drug Regimen")
treatment_outcome_summary_stats = treatment_outcome_summary_stats.merge(final_treatment_tumor_variances, on="Drug Regimen")
treatment_outcome_summary_stats = treatment_outcome_summary_stats.merge(final_treatment_tumor_st_dev, on="Drug Regimen")
treatment_outcome_summary_stats = treatment_outcome_summary_stats.merge(final_treatment_tumor_sem, on="Drug Regimen")

treatment_outcome_summary_stats = treatment_outcome_summary_stats.transpose()
treatment_outcome_summary_stats
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pandas. 
mouse_counts = cleaned_combined_data.groupby(["Drug Regimen", "Timepoint"]).count()

mouse_counts = mouse_counts["Mouse ID"]
mouse_counts.columns = ["Mouse Count"]
mouse_count_plot = mouse_counts.unstack(level=0).plot(kind="bar", figsize=(20, 5))
mouse_count_plot.legend(loc="best")
mouse_count_plot.set_xlim(-0.5, 10.5)
mouse_count_plot.set_title("Number of Mice vs. Time (days)")
mouse_count_plot.set_ylabel("Number of Mice")
mouse_count_plot.set_xlabel("Time (days)")
```

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pyplot.

# We will only look at beginning/end data for this plot, to show an overview of how the treatments worked

overview_mouse_counts = cleaned_combined_data.loc[(cleaned_combined_data["Timepoint"]==0) |
                                                  (cleaned_combined_data["Timepoint"]==45)]
overview_mouse_counts = overview_mouse_counts.groupby(["Drug Regimen", "Timepoint"]).count()

overview_mouse_counts = pd.DataFrame(overview_mouse_counts["Mouse ID"])
overview_mouse_counts.columns = ["Mouse Count"]
overview_mouse_counts.reset_index(inplace=True)
# overview_mouse_counts = overview_mouse_counts.set_index("Drug Regimen")
overview_mouse_counts = overview_mouse_counts.pivot(index=overview_mouse_counts["Drug Regimen"], columns='Timepoint')['Mouse Count']

x_ticks = list(overview_mouse_counts.index)
# y_values = 

fig, ax = plt.subplots()
x_vals = np.arange(len(x_ticks))
width = 0.35

subplt1 = ax.bar(x_vals-width/2, overview_mouse_counts[0], width, label="Starting Mouse Count")
subplt2 = ax.bar(x_vals+width/2, overview_mouse_counts[45], width, label="Ending Mouse Count")



plt.xticks(rotation=90)
ax.set_xticks(x_vals)
ax.set_xticklabels(x_ticks)
ax.set_title("Mouse Counts vs. Drug Regimen")
ax.set_xlabel("Drug Regimen")
ax.set_ylabel("Number of Mice")

plt.legend(loc="best")
plt.show
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas

#show the distribution of female vs. male mice at the beginning of the study here
#start_mice is a subset of the data at timepoint = 0 from earlier
gender_series = start_mice.groupby("Sex").count()
gender_series = gender_series["Mouse ID"]
gender_series.name = ""
print(gender_series)
gender_series.plot(kind="pie", autopct='%1.1f%%', title="Mouse Gender Distribution (Study Start)")
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot

#look at the final gender distribution of the mice at the end of the study
#final_mice is a subset of the data where the timepoint=45 from earlier
gender_series_final = final_mice.groupby("Sex").count()
gender_series_final = gender_series_final["Mouse ID"]
gender_series_final.name = "Mouse Gender Distribution"
print(gender_series_final)
pie_plot, ax_pie = plt.subplots()
plt.pie(gender_series_final, labels=gender_series_final.index, autopct='%1.1f%%')
ax_pie.set_title("Mouse Gender Distribution (Study End)")
```
