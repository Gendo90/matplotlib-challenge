## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
print(mouse_metadata)
print(study_results)

combined_data = study_results.merge(mouse_metadata, on="Mouse ID")

# Display the data table for preview
combined_data
```

```python
# Checking the number of mice.

#check number of mice in original data set
print(len(study_results["Mouse ID"].unique()))
print(len(mouse_metadata["Mouse ID"].unique()))

#show that the number of mice in the combined data set is the same
print(len(combined_data["Mouse ID"].unique()))
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

#first show how many values there are for each timepoint - that is, how many mice have a given number of timepoints?
grouped_by_id = study_results.groupby("Mouse ID")
check_duplicates = grouped_by_id.count()

check_duplicates = check_duplicates["Timepoint"]
check_duplicates

per_value = check_duplicates.value_counts()
print(per_value)

#group all the mice by id and then by timepoint, so that we can count how many timepoints for each mouse there are
#to see if there are duplicate timepoints for the same mouse (e.g. the count is greater than one for at least one timepoint
#for a particular mouse)
timepoint_duplicates = study_results.groupby(["Mouse ID", "Timepoint"]).count()
timepoint_duplicates =timepoint_duplicates.drop(columns="Tumor Volume (mm3)")
timepoint_duplicates.columns = ["Timepoint Count"]
print(timepoint_duplicates)
timepoint_duplicates = timepoint_duplicates.loc[timepoint_duplicates["Timepoint Count"]>1]


timepoint_duplicates.head()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse = combined_data.loc[combined_data["Mouse ID"]=="g989"]
duplicate_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_combined_data = combined_data.loc[combined_data["Mouse ID"]!="g989", :]
cleaned_combined_data
```

```python
# Checking the number of mice in the clean DataFrame.
print(len(cleaned_combined_data["Mouse ID"].unique()))
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.

#this summary statistics table shows data for the mice at the beginning of the study, at timepoint=0

#filter table so that it only shows the remaining mice and tumor values at timepoint 45
start_mice = cleaned_combined_data.loc[cleaned_combined_data["Timepoint"]==0]
print(f"There are {len(start_mice)} mice at the start of the study.")

index_values = ["Mean", "Median", "Variance", "Standard Deviation", "SEM"]

#write a function that computes the required values for each drug regimen and puts it in a series
def statsForTreatment(treatment, dataset):
    treatment_subset = dataset.loc[dataset["Drug Regimen"]==treatment]
    #get mean tumor volume
    tumor_mean = treatment_subset["Tumor Volume (mm3)"].mean()
    #get median tumor volume
    tumor_median = treatment_subset["Tumor Volume (mm3)"].median()
    #get variance in tumor volume
    tumor_variance = treatment_subset["Tumor Volume (mm3)"].var()
    #get standard deviation in tumor volume
    tumor_st_dev = treatment_subset["Tumor Volume (mm3)"].std()
    #get sem in tumor volume
    tumor_sem = treatment_subset["Tumor Volume (mm3)"].sem()
    
    output_df = pd.DataFrame({treatment:[tumor_mean, tumor_median, tumor_variance, tumor_st_dev, tumor_sem]}, index=index_values)
    return output_df

all_treatments = list(cleaned_combined_data["Drug Regimen"].unique())
print(all_treatments)

summary_stats_iterative = statsForTreatment(all_treatments[0], cleaned_combined_data)
summary_stats_iterative.index.rename = "Stat Variables"

for treatment in all_treatments[1:]:
    new_stats = statsForTreatment(treatment, start_mice)
    new_stats.index.rename = "Stat Variables"
    summary_stats_iterative = summary_stats_iterative.merge(new_stats, left_index = True, right_index = True)
    
summary_stats_iterative
    
    
    
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function

#this summary statistics table shows the tumor volume for the mice by treatment at the end of the study (timepoint = 45)
final_mice = cleaned_combined_data.loc[cleaned_combined_data["Timepoint"]==45]
print(f"There are {len(final_mice)} mice left at the end of the study.")


#group by drug regimen to get summary stats
grouped_treatments = final_mice.groupby("Drug Regimen")

#calculate the mean across all treatments
final_treatment_tumor_means = pd.DataFrame(grouped_treatments.mean()["Tumor Volume (mm3)"])
final_treatment_tumor_means.columns = ["Mean"]

#set the summary stats dataframe to be used for the final mice statistics
treatment_outcome_summary_stats = final_treatment_tumor_means

#calculate the median across all treatments
final_treatment_tumor_medians = pd.DataFrame(grouped_treatments.median()["Tumor Volume (mm3)"])
final_treatment_tumor_medians.columns = ["Median"]

#calculate the variance across all treatments
final_treatment_tumor_variances = pd.DataFrame(grouped_treatments.var()["Tumor Volume (mm3)"])
final_treatment_tumor_variances.columns = ["Variance"]

#calculate the standard deviation across all treatments
final_treatment_tumor_st_dev = pd.DataFrame(grouped_treatments.std()["Tumor Volume (mm3)"])
final_treatment_tumor_st_dev.columns = ["Standard Deviation"]

#calculate the sem across all treatments
final_treatment_tumor_sem = pd.DataFrame(grouped_treatments.sem()["Tumor Volume (mm3)"])
final_treatment_tumor_sem.columns = ["SEM"]

#add all the calculated columns to the final summary stats table
treatment_outcome_summary_stats = treatment_outcome_summary_stats.merge(final_treatment_tumor_medians, on="Drug Regimen")
treatment_outcome_summary_stats = treatment_outcome_summary_stats.merge(final_treatment_tumor_variances, on="Drug Regimen")
treatment_outcome_summary_stats = treatment_outcome_summary_stats.merge(final_treatment_tumor_st_dev, on="Drug Regimen")
treatment_outcome_summary_stats = treatment_outcome_summary_stats.merge(final_treatment_tumor_sem, on="Drug Regimen")

treatment_outcome_summary_stats = treatment_outcome_summary_stats.transpose()
treatment_outcome_summary_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
mice_final_timepoint = cleaned_combined_data.loc[cleaned_combined_data["Timepoint"]==45]

interesting_treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
remaining_mice_interesting_treatments = mice_final_timepoint.loc[mice_final_timepoint["Drug Regimen"].isin(interesting_treatments)]
remaining_mice_interesting_treatments
```

```python
# Put treatments into a list for for loop (and later for plot labels)


# Create empty list to fill with tumor vol data (for plotting)
tumor_volume_plot_data = None


# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for treatment in interesting_treatments:
    
#     Locate the rows which contain mice on each drug and get the tumor volumes
    mice_in_treatment = remaining_mice_interesting_treatments.loc[
        remaining_mice_interesting_treatments["Drug Regimen"]==treatment]
    tumor_volumes = mice_in_treatment["Tumor Volume (mm3)"]
    
    #put into plotting dataframe
    col_label = treatment
    this_treatment = pd.DataFrame(tumor_volumes)
    this_treatment.columns = [col_label]

    if(tumor_volume_plot_data is not None):
        tumor_volume_plot_data = tumor_volume_plot_data.merge(this_treatment, left_index=True, right_index=True, how="outer")
    else:
        tumor_volume_plot_data = this_treatment
    
    #calculate quartile statistics
    remaining_median = tumor_volumes.median()
    remaining_lower_quartile = np.percentile(tumor_volumes, 25)
    remaining_upper_quartile = np.percentile(tumor_volumes, 75)
    
    #output quartiles
    print("The current treatment being examined is ", treatment)
    print("The first, second (median), and third quartiles are {:.2f}, {:.2f}, and {:.2f}"
          .format(remaining_lower_quartile, remaining_median, remaining_upper_quartile))

    #calculate IQR
    remaining_interquartile_range = remaining_upper_quartile-remaining_lower_quartile

    lower_outlier_bound = remaining_lower_quartile-1.5*(remaining_interquartile_range)
    upper_outlier_bound = remaining_upper_quartile+1.5*(remaining_interquartile_range)
    
    #output IQR and lower/upper quartile boundaries
    print("The calculated IQR is {:.2f}".format(remaining_interquartile_range))
    print("THe lower and upper outlier bounds are {:.2f} and {:.2f}".format(lower_outlier_bound, upper_outlier_bound))

    
    lower_outliers = mice_in_treatment.loc[mice_in_treatment["Tumor Volume (mm3)"]<=lower_outlier_bound]
    if(not len(lower_outliers) == 0):
        print("The lower outliers are: ", lower_outliers)
    upper_outliers = mice_in_treatment.loc[mice_in_treatment["Tumor Volume (mm3)"]>=upper_outlier_bound]
    if(not len(upper_outliers) == 0):
        print("The upper outliers are: ", upper_outliers)
        
    if(len(upper_outliers)==len(lower_outliers)==0):
        print("No outliers for this treatment.")
    
    #line space break before next treatment output
    print(" ")
    
    
    # Determine outliers using upper and lower bounds
tumor_volume_plot_data
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
#use tumor_volume_plot_data dataframe to plot data

tumor_boxplot = tumor_volume_plot_data.boxplot()

tumor_boxplot.set_title("Final Tumor Volumes (mm3) by Treatment")
tumor_boxplot.set_ylabel("Tumor Volume (mm3)")
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
import random

#we want a random mouse to be shown, so we can use the random module to select a mouse from the list of mice treated by 
#Capomulin


#first, get the subset of Capomulin mice
capomulin_mice = cleaned_combined_data.loc[cleaned_combined_data["Drug Regimen"]=="Capomulin"]

#now get the random mouse
possible_mice = list(capomulin_mice["Mouse ID"].unique())
selected_mouse_ID = possible_mice[random.randint(0, len(possible_mice)-1)]

#get data for specific mouse
selected_mouse_data = capomulin_mice.loc[capomulin_mice["Mouse ID"]==selected_mouse_ID]

#plot data for that mouse
capomulin_mouse_plot, ax_capomulin_mouse = plt.subplots()

plt.plot(selected_mouse_data["Timepoint"], selected_mouse_data["Tumor Volume (mm3)"])

ax_capomulin_mouse.set_title(f"{selected_mouse_ID} Tumor Volume (mm3) over Time (days)")
ax_capomulin_mouse.set_xlabel("Time (days)")
ax_capomulin_mouse.set_ylabel("Tumor Volume (mm3)")
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen

#first get the mouse weight and average tumor volume for each mouse
avg_tumors = capomulin_mice.groupby("Mouse ID").mean()
capomulin_mice_weights = avg_tumors["Weight (g)"]
avg_tumors = avg_tumors["Tumor Volume (mm3)"]

capulin_scatterplot, ax_scatter = plt.subplots()

plt.scatter(capomulin_mice_weights, avg_tumors)

ax_scatter.set_title("Capomulin Mice Weights (g) vs. Avg. Tumor Volumes (mm3)")
ax_scatter.set_xlabel("Avg. Tumor Volumes (mm3)")
ax_scatter.set_ylabel("Mice Weights (g)")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

#use linregression from scipy
(slope, y_intercept, corr, _, _) = st.linregress(capomulin_mice_weights, avg_tumors)

sorted_weights = sorted(capomulin_mice_weights.values)
best_fit_line_x = np.array([a for a in range(int(sorted_weights[0]), int(sorted_weights[-1])+1)])
best_fit_line = slope*best_fit_line_x+y_intercept


capulin_lin_fit_scatterplot, ax_lin_fit_scatter = plt.subplots()

plt.scatter(capomulin_mice_weights, avg_tumors, label="Raw Data")
plt.plot(best_fit_line_x, best_fit_line, "r--", label="Best Fit Line (r=0.84)")

ax_lin_fit_scatter.set_title("Capomulin Mice Weights (g) vs. Avg. Tumor Volumes (mm3)")
ax_lin_fit_scatter.set_xlabel("Avg. Tumor Volumes (mm3)")
ax_lin_fit_scatter.set_ylabel("Mice Weights (g)")
plt.legend(loc="best")

print("The correlation (r) value for a linear fit of the mice weights vs. tumor volumes for the capomulin mice is {:.2f}"
      .format(corr))

print("The resulting best linear fit line, as compared to the scatter plot, is show below:")
```
